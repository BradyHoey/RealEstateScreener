```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send an HTTP GET request to the website
url = "https://www.bankrate.com/insurance/homeowners-insurance/states/#state"
response = requests.get(url)
html_content = response.content

# Parse the HTML content
soup = BeautifulSoup(html_content, "html.parser")

# Find the table element that contains the data
table_element = soup.find("table", class_="table-inline")

# Find all rows in the table except the header row
rows = table_element.find_all("tr")[1:]

# Extract the state names and corresponding costs
data = []
for row in rows:
    cells = row.find_all("td")
    state_name = cells[0].text.strip()
    cost = cells[1].text.strip("$")
    data.append({"State": state_name, "Cost": cost})

# Create a DataFrame from the extracted data
df = pd.DataFrame(data)

# Export the DataFrame to a CSV file
output_file = "homeowners_insurance_cost.csv"
df.to_csv(output_file, index=False)
```
